# Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the dataset

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
print(df.columns.values)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
mean_ages = df.groupby(['sex','pclass'])['age'].mean()
display(mean_ages)
```

```python
def replace_nan_age(row):
    if pd.isnull(row['age']):
        return mean_ages[row['sex'], row['pclass']]
    else:
        return row['age']
    
df['age'] = df.apply(replace_nan_age, axis=1)
df.isnull().sum()
```

```python
df['age_bin'] = pd.cut(df['age'], bins=[0, 18, 30, 50, 100], labels=['Child', 'Teen', 'Adult', 'Old'])
age_group_survival = df.groupby('age_bin')['survived'].mean().reset_index()
```

```python
# create a bar plot of the survival rate by age group
sns.barplot(x='age_bin', y='survived', data=age_group_survival)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()
```
