```python
# import libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from textblob import TextBlob
import pandas as pd
nltk.download('punkt')
nltk.download('wordnet')
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

df = pd.read_csv("spam.csv", encoding="latin-1")
df.head()
```

```python
# Remove unnecessary columns and rename remaining columns
df = df.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
df = df.rename(columns={"v1":"label", "v2":"text"})
df.head()
```

```python
df.describe()
```

```python
df.groupby("label").describe()
```

```python
# labeled spam(1) and ham(0) 
df['spam'] = df['label'].map( {'spam': 1, 'ham': 0} ).astype(int)
df.head()
```

```python
# calculate length
df['length'] = df['text'].apply(len)
df.head()
```

```python
# Remove punctuation and convert text to lowercase
import string
df["text"] = df["text"].apply(lambda x: x.translate(str.maketrans("", "", string.punctuation)))
df["text"] = df["text"].apply(lambda x: x.lower())
df.head()
```

```python
# Tokenize text
df["tokens"] = df["text"].apply(nltk.word_tokenize)
df.head()
```

```python
# Remove stop words
nltk.download('stopwords')
stop_words = set(stopwords.words("english"))
df["no_stopwords"] = df["tokens"].apply(lambda x: [word for word in x if word not in stop_words])
df.head()
```

```python
# Stem words
porter_stemmer = PorterStemmer()
df["stemmed"] = df["no_stopwords"].apply(lambda x: [porter_stemmer.stem(word) for word in x])
df.head()
```

```python
# Lemmatize words
wordnet_lemmatizer = WordNetLemmatizer()
df["lemmatized"] = df["no_stopwords"].apply(lambda x: [wordnet_lemmatizer.lemmatize(word) for word in x])
df.head()
```

```python
from textblob import TextBlob
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# perform POS tagging on words in 'new' column
df['PoS'] = df['lemmatized'].apply(lambda x: TextBlob(" ".join(x)).tags)

df.head()
```

```python
# calculating TF-IDF
import numpy as np
# Calculate term frequency-inverse document frequency (TF-IDF)
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df["text"])
tfidf_matrix
np.shape(tfidf_matrix)
```

```python
import pandas as pd
import nltk

# Assuming you have loaded the dataframe 'df' with the necessary data
# Step 9: Calculating TF-IDF
N = len(df)
def doc_freq(word , df):
    df = sum(1 for text in df['lemmatized'] if word in text)
    return df

def calc_tf_idf(doc, token, df):
    words_count = len(df['lemmatized'][doc])
    counter = dict(nltk.FreqDist(df['lemmatized'][doc]))
    tf = counter[token] / words_count
    df = doc_freq(token , df)
    idf = np.log(N/(df+1))
    tf_idf = tf*idf
    return tf_idf

# Iterate over each document in the 'lemmatized' column
for doc in range(len(df['lemmatized'])):
    # Iterate over each token in the lemmatized document
    for token in df['lemmatized'][doc]:
        tf_idf = calc_tf_idf(doc=doc, token=token, df=df)
        print(token," ", tf_idf)
```
